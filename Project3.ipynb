* You loop through a set using a FOR-loop like this "for item in set:"
* The less-than sign means "strict subset"... so S1 < S2 means every item in S1
is in S2, but there are still extra items in S2 that are not in S1.
* You can use "item = s.pop()" to remove one item from a set and place it in item. If
you want to replace it, you can use s.add(item) --- but pop is intended to destroy the
